```python
%pip install plotly
%pip install nbformat
import pandas as pd
import plotly.express as px

# 1. Load the CSV you and Kyle made
df = pd.read_csv('desi_catalog.csv')

# 2. CRITICAL: Sample the data 
# Web browsers will crash if you give them millions of points.
# A random 100,000 galaxies to see the shape of the cosmic web. (I tried doing the whole thing and vscode just gave up. Add more at your own risk.)
df_sample = df.sample(n=100000)

# 3. Create the 3D interactive plot
print("Rendering universe...")
fig = px.scatter_3d(
    df_sample, 
    x='x', y='y', z='z', 
    color='object_type', # This will color-code them by Galaxy, QSO, etc.
    opacity=0.7
)

# Make the dots smaller so they look like stars/galaxies
fig.update_traces(marker=dict(size=2))

# Remove the white background to make it look more like space
fig.update_layout(template="plotly_dark")

fig.show()
```
